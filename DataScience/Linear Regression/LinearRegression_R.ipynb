```r
install.packages("caTools")
set.seed(123)
```

```r
library(caTools)
dataset = read.csv("Salary_Data.csv")
split.data = sample.split(dataset, SplitRatio=2/3)
test.data = subset(dataset, FALSE)
train.data = subset(dataset, TRUE)

regression = lm(Salary ~ YearsExperience, train.data)
y.predict = predict(regression, test.data)
summary(regression)
```

```r
install.packages("ggplot2")
library(ggplot2)
```

```r
ggplot() +
geom_point(aes(train.data$YearsExperience, train.data$Salary), color="Red") +
geom_line(aes(test.data$YearsExperience, y.predict), color="Blue")
```
