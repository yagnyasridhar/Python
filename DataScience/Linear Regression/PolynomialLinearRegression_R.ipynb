```r
install.packages("ggplot2")
```

```r
library(ggplot2)
dataset = read.csv("Position_Salaries.csv")
```

```r
dataset = dataset[2:3]
```

```r
linear = lm(dataset$Salary ~., dataset)
yPred = predict(linear, data.frame(Level = 6.5))
```

```r
yPred
```

```r
dataset$Level4 = (dataset$Level)^4
dataset$Level2 = (dataset$Level)^2
dataset$Level3 = (dataset$Level)^3
```

```r
dataset
```

```r
linear_poly = lm(dataset$Salary ~., dataset)
yPred1 = predict(linear_poly, data.frame(Level = 6.5, Level4 = (6.5)^4, Level2 = (6.5)^2, Level3 = (6.5)^3))
```

```r
yPred1
```

```r
ggplot() +
geom_point(aes(dataset$Level, dataset$Salary), color="Red") +
geom_line(aes(dataset$Level, predict(linear)), color="Blue")+
ggtitle("Salary Prediction")+
xlab("Level")+
ylab("Salary")
```

```r
ggplot() +
geom_point(aes(dataset$Level4, dataset$Salary), color="Red") +
geom_line(aes(dataset$Level4, predict(linear_poly)), color="Blue")+
ggtitle("Salary Prediction")+
xlab("Level")+
ylab("Salary")
```
