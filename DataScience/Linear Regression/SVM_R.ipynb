```r
install.packages("e1071")
```

```r
library(e1071)
```

```r
dataset = read.csv("Position_Salaries.csv")
dataset = dataset[2:3]
```

```r
regression = svm(dataset$Salary ~ ., dataset, type = "eps-regression")
```

```r
yPred = predict(regression, data.frame(Level=6.5))
yPred
```

```r
install.packages("ggplot2")
```

```r
library(ggplot2)
```

```r
ggplot() +
geom_point(aes(dataset$Level, dataset$Salary),color="red") +
geom_line(aes(dataset$Level, predict(regression,dataset)),color="blue")
```
